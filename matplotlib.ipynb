```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.plot(x,y)
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.plot(x,y, marker='o', linestyle='--',color='red')
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.bar(x,y,color='red')
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.scatter(x,y,color='red')
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.hist(y, bins=5)
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.hist(y, bins=15)
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [10,20,30,40,50,60,70]

plt.hist(y, bins=15,color='orange',edgecolor='black')
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
subjects=["Eng","Hindi","comp","maths","science"]
score=[89,30,67,89,88]


plt.pie(score,labels=subjects)
plt.title("plotting")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
